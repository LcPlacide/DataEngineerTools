```python
# Chargement du json
dtype={'duration':'datetime64[ns]'}
anime_df=pd.read_json("animescrawler/animescrawler/spiders/anime_complet2.json",dtype=dtype)

# Suppression des éventuels doublons
col=anime_df.columns.to_list()
col.remove("duration")
anime_df=anime_df.loc[anime_df[col].astype(str).drop_duplicates().index]
anime_df.synopsis.value_counts()

# Rajout des NaN
NA=['','None',"N/A",'Unknown',None,[],dict(),set()]
for col in anime_df.columns:
    anime_df[col]=anime_df[col].apply(lambda x: np.NaN if x in NA else x)
anime_df.head(2)
anime_df.info()
```
